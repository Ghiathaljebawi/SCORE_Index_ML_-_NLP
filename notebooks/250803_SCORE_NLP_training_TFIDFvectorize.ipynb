```python
# split data
X, y = dataset["text"], dataset["sco2 - Actor\'s relevance"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 30)
```

```python
# Tokenize
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

X_train_tok = X_train.apply(lambda t: word_tokenize(str(t).lower()))
X_test_tok  = X_test.apply(lambda t: word_tokenize(str(t).lower()))

X_train_tok.head()
```

```python
# Lemmatize
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer   = WordNetLemmatizer()

X_train_lem = X_train_tok.apply(lambda lst: [lemmatizer.lemmatize(w) for w in lst])
X_test_lem = X_test_tok.apply(lambda lst: [lemmatizer.lemmatize(w) for w in lst])

X_train_lem.head()
```

```python
# Remove stopwords
from nltk.corpus import stopwords
nltk.download('stopwords')

stops = set(stopwords.words('english'))

X_train_nostop = X_train_lem.apply(lambda lst: [w for w in lst if w not in stops])
X_test_nostop = X_test_lem.apply(lambda lst: [w for w in lst if w not in stops])

X_train_nostop.head(100)
```

```python
# Keep words > 2
import re
X_train_filt = X_train_nostop.apply(lambda lst: [w for w in lst if len(w) > 2])
X_test_filt = X_test_nostop.apply(lambda lst: [w for w in lst if len(w) > 2])

X_train_filt.head()
```

```python
# Verctorize
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. join tokens back to strings
train_clean = X_train_filt.apply(' '.join)
test_clean  = X_test_filt.apply(' '.join) 

# 2. count vect
# vect = CountVectorizer()
vect = TfidfVectorizer()
X_train_vec = vect.fit_transform(train_clean)
X_test_vec  = vect.transform(test_clean)

print("train matrix shape:", X_train_vec.shape)
```

```python
X_train = X_train_vec
X_test = X_test_vec
```

XGBoost Regressor:

```python
# XGBoost Regressor + Grid Search + Metrics

from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

xgb = XGBRegressor(objective="reg:squarederror", random_state=30)

xgb_grid = {
    "n_estimators":      [200, 500, 800, 1100],       # more trees to support lower learning rate
    "learning_rate":     [0.01, 0.05, 0.1, 0.3],      # smaller steps → less overfit, better generalisation
    "max_depth":         [3, 4, 6, 9],              # control tree complexity
    #"min_child_weight":  [1, 3, 5],              # minimum sum-hessian in child → combats overfit
    "subsample":         [0.8, 1.0],             # row sampling; <1.0 adds randomness
    "colsample_bytree":  [0.8, 1.0],             # column sampling per tree
    #"gamma":             [0, 0.1, 0.3],          # min loss-reduction to make a split
    #"reg_alpha":         [0, 0.1],               # L1 regularisation
    #"reg_lambda":        [1, 3],                 # L2 regularisation
}

xgb_cv = GridSearchCV(
    xgb,
    xgb_grid,
    cv=5,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1,
    verbose=0,
)
xgb_cv.fit(X_train, y_train)
best_xgb = xgb_cv.best_estimator_

y_pred = best_xgb.predict(X_test)
mse  = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae  = mean_absolute_error(y_test, y_pred)
r2   = r2_score(y_test, y_pred)

print("XGB best params:", xgb_cv.best_params_)
print(f"XGB  →  MSE: {mse:.3f} | RMSE: {rmse:.3f} | MAE: {mae:.3f} | R²: {r2:.3f}")
```

Random Forest Regressor:

```python
# Random Forest Regressor + Grid Search + Metrics

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

rf = RandomForestRegressor(random_state=30)

rf_grid = {
    "n_estimators": [300, 500, 800],     # trees — enough to stabilise, not crazy-high
    "max_depth":         [10, 20, 30],   # depth / size constraints - we can use instead: "max_leaf_nodes": [500, 1000]
    "max_features":      ["sqrt", 0.3, 0.5],    # sqrt ≈ √p, or explicit fractions
    "min_samples_split": [2, 5, 10],     # regularisation: make splits less eager
    "min_samples_leaf":  [1, 2, 4],
    "bootstrap":         [True, False],     # sub-sampling entire rows
}

rf_cv = GridSearchCV(
    rf,
    rf_grid,
    cv=5,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1,
    verbose=0,
)
rf_cv.fit(X_train, y_train)
best_rf = rf_cv.best_estimator_

y_pred_rf = best_rf.predict(X_test)
mse_rf  = mean_squared_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
mae_rf  = mean_absolute_error(y_test, y_pred_rf)
r2_rf   = r2_score(y_test, y_pred_rf)

print("RF best params:", rf_cv.best_params_)
print(f"RF   →  MSE: {mse_rf:.3f} | RMSE: {rmse_rf:.3f} | MAE: {mae_rf:.3f} | R²: {r2_rf:.3f}")
```
